```python
# Clean sentences
english_sentence = [clean_sentence(sentence) for sentence in english_sentence]
hindi_sentence = [clean_sentence(sentence) for sentence in hindi_sentence]

# Tokenize words
eng_text_tokenized, eng_text_tokenizer = tokenize(english_sentence)
hin_text_tokenized, hin_text_tokenizer = tokenize(hindi_sentence)

max_english_len = len(max(eng_text_tokenized, key=len))
max_hindi_len = len(max(hin_text_tokenized, key=len))

print("Maximum length english sentence: ", max_english_len)
print("Maximum length hindi sentence: ", max_hindi_len)

# Check language length
english_vocab = len(eng_text_tokenizer.word_index) + 1
hindi_vocab = len(hin_text_tokenizer.word_index) + 1

print("English vocabulary is of {} unique words".format(english_vocab))
print("Hindi vocabulary is of {} unique words".format(hindi_vocab))
```

```python
english_pad_sentence = pad_sequences(eng_text_tokenized, max_english_len, padding="post")
hindi_pad_sentence = pad_sequences(hin_text_tokenized, max_hindi_len, padding="post")

english_pad_sentence[0]
```

```python
english_pad_sentence = english_pad_sentence.reshape(*english_pad_sentence.shape, 1)
hindi_pad_sentence = hindi_pad_sentence.reshape(*hindi_pad_sentence.shape, 1)

english_pad_sentence[0]
```

## Encoder-Decoder

```python
input_sequence = Input(shape=(max_english_len))
embedding = Embedding(input_dim=english_vocab, output_dim=128, )(input_sequence)
encoder = LSTM(64, return_sequences=False)(embedding)

r_vec = RepeatVector(max_hindi_len)(encoder)
decoder = LSTM(64, return_sequences=True, dropout=0.2)(r_vec)

logits = TimeDistributed(Dense(hindi_vocab))(decoder)
```

```python
model = Model(input_sequence, Activation('softmax')(logits))
model.compile(loss=sparse_categorical_crossentropy, optimizer=Adam(1e-3), metrics=['accuracy'])
model.summary()
```

```python
modelResults = model.fit(english_pad_sentence, hindi_pad_sentence, batch_size=30, epochs=8)
```

```python
def logits_to_sentence(logits, tokenizer):
    index_to_words = {idx: word for word, idx in tokenizer.word_index.items()}
    index_to_words[0] = '<empty>'
    #print([index_to_words[prediction] for prediction in np.argmax(logits, 1)])
    
    
    print(np.argmax(logits, 1))
    
    return ' '.join([index_to_words[prediction] for prediction in np.argmax(logits, 1)])

index = 120
print("The english sentence: ", english_sentence[index])
print("The hindi sentence: ", hindi_sentence[index])
print("")
print("The predicted sentence: ", end="")
print(logits_to_sentence(model.predict(english_pad_sentence[index:index+1])[0], hin_text_tokenizer))

print("------------------------------------------------------------------------------------------------")
index = 124
print("The english sentence: ", english_sentence[index])
print("The hindi sentence: ", hindi_sentence[index])
print("")
print("The predicted sentence: ", end="")
logits_to_sentence(model.predict(english_pad_sentence[index:index+1])[0], hin_text_tokenizer)
```
